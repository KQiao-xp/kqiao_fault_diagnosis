```python
import scipy.io as scio # To use the '.mat' files
import seaborn as sns
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
for root, dirs, files in os.walk("data483", topdown=False):
    for file_name in files:
        path = os.path.join(root, file_name) # 路径拼接
        print(path)
```

```python
# 正常状态
mat = scio.loadmat(r'data483\Normal_3.mat')
key_name = list(mat.keys())[3]
DE_data = mat.get(key_name)
df = pd.DataFrame(DE_data)

plt.figure(figsize=(10,1))
plt.title('Raw vibration signal for healthy state')
plt.ylabel('Amplitude')
# plt.xlabel('Data length')
plt.ylim((-2, 2))
plt.plot(df.iloc[0:2000,0])
plt.show()

index = np.arange(0, 2000, 8)
plt.figure(figsize=(10,1))
plt.title('Preprocessed vibration signal for healthy state')
# plt.xlabel('Data length')
plt.ylabel('Amplitude')
plt.ylim((-2, 2))
plt.plot(DE_data[index])
plt.show()
```
